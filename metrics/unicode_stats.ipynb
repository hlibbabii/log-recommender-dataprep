```python
from dataprep.util import read_dict_from_2_columns
from dataprep.noneng import is_non_eng
from typing import Dict

path_to_nosplit_vocab = '/home/lv71161/hlibbabii/nosplitvocab/vocab'

all_vocab = read_dict_from_2_columns(path_to_nosplit_vocab)
print(f"Total vocab size: {len(all_vocab)}")
```

```python
def get_all_non_ascii(all_vocab: Dict[str, int]) -> Dict[str, int]:
    return {k: v for k, v in all_vocab.items() if is_non_eng(k)}


noneng_vocab = get_all_non_ascii(all_vocab)
print(f"Non-eng vocab size: {len(noneng_vocab)}")
```

```python
import random
n=1000
selected_non_eng = random.choices(list(noneng_vocab.items()), k=n)
print(f"Selected {n} non-eng tokens: {selected_non_eng}")
```

```python
from dataprep.noneng import is_non_eng

def probably_real_noneng_word(word: str, n_chars: int=1, min_len:int=6):
        return sum(map(lambda ch: int(is_non_eng(ch)), word)) < (len(word)-1) // 3

probably_real_noneng_vocab = {word: freq for word, freq in noneng_vocab.items() if probably_real_noneng_word(word)}
print(f"Size of probably real non-eng vocab\n (the ratio of 'strange' chars to all chars in the word is not high. 7-9 - 1 strange char; 10-12: 2 etc.'): {len(probably_real_noneng_vocab)}")
```

```python
import random
n=1000
selected_prob_real_non_eng = random.choices(list(probably_real_noneng_vocab.items()), k=n)
print(f"Selected {n} probably real non-eng tokens: {selected_prob_real_non_eng}")
```

```python
def probably_n_words(word, n, min_len_per_word:int=4):
    return len(word) >= (min_len_per_word * n) and (n-1) == sum(map(lambda ch: int(ch.isupper() or ch == '_'), word[1:]))

def probably_2_to_5_words(word):
    return probably_n_words(word, 2) or probably_n_words(word, 3) or probably_n_words(word, 4) or probably_n_words(word, 5)

probably_2to5_words_real_noneng_vocab = {word: freq for word, freq in probably_real_noneng_vocab.items() if probably_2_to_5_words(word)}
print(f"All probably real non-eng tokens probably containing 2-5 words ({len(probably_2to5_words_real_noneng_vocab)}):")
print(probably_2to5_words_real_noneng_vocab)
```

```python
import dataprep
import pprint
from spiral import ronin
import logging
logging.getLogger('Ronin').setLevel(logging.INFO)

ronin_non_eng = [ronin.split(non_eng) for non_eng, _ in probably_2to5_words_real_noneng_vocab.items()]
dataprep_non_eng = [dataprep.basic(non_eng) for non_eng, _ in probably_2to5_words_real_noneng_vocab.items()]
print("Probably real non-eng tokens probably containing 2-5 words split by Ronin alg (spiral) and our basic parser:")
pprint.pprint([*zip(ronin_non_eng, dataprep_non_eng)])
```
